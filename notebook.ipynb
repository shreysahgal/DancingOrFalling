```python
formats = info_dict.get('formats', None)
```

```python
def process_video_parallel(url, frames_per_proc, proc_id, num_imgs=5, save_path="/Users/shreysahgal/Documents/dance_fall_data_collector/data"):
    print(f"hello from proc {proc_id}")
    
    cap = cv.VideoCapture(url)
    cap.set(cv.CAP_PROP_POS_FRAMES, frames_per_proc * proc_id)

    img_count = 0
    frame_count = 0
    
    while img_count < num_imgs and frame_count < frames_per_process:
        print(f"proc {proc_id} processing img {img_count}...")
        ret, frame = cap.read()

        if not ret:
            break

        fname = f"{str(proc_id)}_{str(img_count)}.png"
        cv.imwrite(os.path.join(save_path, fname), frame)
        
        img_count += 1
        frame_count += (frames_per_proc // num_imgs)
        
        cap.set(1, frame_count)
    cap.release()
    
```
